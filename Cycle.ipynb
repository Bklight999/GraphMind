```python
import json

with open('./data_0104/Cycle_detection.json','r') as f:
    datas = json.load(f)

print(datas[0]['result'])
```

```python
print(datas[0]['result'])
print(datas[0].keys())
edges = datas[0]['edges']
edges = [tuple(edge) for edge in edges]
print(edges)
```

```python
final_datas = []
for data in datas:
    edges = data['edges']
    edges = [tuple(edge) for edge in edges]
    node_num = data['node_num']
    edges_str = ', '.join([f"({i}, {j})" for i, j in edges])
    question = f"""Determine whether or not there is a cycle in an undirected graph. 
In an undirected graph, (i,j) means that node i and node j are connected with an undirected edge. 
Given a graph, you need to output Yes or No, indicating whether there is a cycle in the graph.
Q: The nodes are numbered from 0 to {node_num-1}, and the edges are: {edges_str}. Is there a cycle in this graph?" """

    answers = data['result']
    output = """"""
    if data['result'] == [-1]:
        hint = f"Hint: There is no cycle in this graph.\n"
        requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [No].\n'
        final_datas.append({
            'id': len(final_datas),
            'input': question,
            'hint': hint,
            'requirement': requirement
        })
    else:
        for answer in answers:
            hint = f"Hint: A cycle in this graph is as follows:"
            answer.append(answer[0])
            path = ' -> '.join([str(node) for node in answer])
            hint += f"\n{path}\n"
            requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [Yes].\n'
            final_datas.append({
                'id': len(final_datas),
                'input': question,
                'hint': hint,
                'requirement': requirement
            })
```
