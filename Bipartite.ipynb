```python
import json

with open('./data_0104/Bipartite.json','r') as f:
    datas = json.load(f)

print(datas[0]['result'])
```

```python
print(datas[0]['result'])
print(datas[0].keys())
edges = datas[0]['edges']
edges = [tuple(edge) for edge in edges]
print(edges)
```

```python
final_datas = []
for data in datas:
    edges = data['edges']
    edges = [tuple(edge) for edge in edges]
    node_num = data['node_num']
    edges_str = ', '.join([f"({i}, {j})" for i, j in edges])
    question = f"""Determine whether or not a graph is bipartite. 
  In a directed graph, (i->j) means that node i and node j are connected with an directed edge from node i to node j.
  Given a graph, you need to output Yes or No, indicating whether the graph is bipartite. 
  Q: The nodes are numbered from 0 to {node_num-1}, and the edges are: {edges_str}. Is this graph bipartite? """

    answers = data['path']
    output = """"""
    if data['result'] == True:
        hint = f"Hint: The given graph is a bipartite graph, with two node sets: {data['path'][0]} and {data['path'][1]}.\n"
        requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [Yes].\n'
        final_datas.append({
            'id': len(final_datas),
            'input': question,
            'hint': hint,
            'requirement': requirement
        })
    else:
        for answer in answers:
            hint = f"Hint: The given graph is not a bipartite graph, since node {answer[0]} and node {answer[1]} are in the same set.\n"
            # answer.append(answer[0])
            # path = ' -> '.join([str(node) for node in answer])
            # hint += f"\n{path}\n"
            requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [No].\n'
            final_datas.append({
                'id': len(final_datas),
                'input': question,
                'hint': hint,
                'requirement': requirement
            })
```
