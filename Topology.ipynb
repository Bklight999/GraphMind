```python
import json

with open('./data_0104/Topological_sort.json','r') as f:
    datas = json.load(f)

print(datas[0]['result'])
```

```python
print(datas[0]['result'])
print(datas[0].keys())
edges = datas[0]['edges']
edges = [tuple(edge) for edge in edges]
print(edges)
```

```python
final_datas = []
for data in datas:
    edges = data['edges']
    edges = [tuple(edge) for edge in edges]
    node_num = data['node_num']
    edges_str = ', '.join([f"({i}, {j})" for i, j in edges])
    question = f"""Find one of the topology sorting paths of the given graph.
In a directed graph, (i->j) means that node i and node j are connected with a directed edge from node i to node j. 
Given a graph, you need to output one of the topology sorting paths of the graph.
Q: The nodes are numbered from 0 to {node_num-1}, and the edges are: {edges_str}. Give one topology sorting path of this graph."""

    answers = data['result']
    output = """"""
    if data['result'] == [-1]:
        hint = f"Hint: There is no topological sort in this graph.\n"
        requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [None].\n'
        final_datas.append({
            'id': len(final_datas),
            'input': question,
            'hint': hint,
            'requirement': requirement
        })
    else:
        for answer in answers:
            hint = f"Hint: A topology sorting path in this graph is as follows:"
            path = ' -> '.join([str(node) for node in answer])
            hint += f"\n{path}\n"
            requirement = f'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [{path}].\n'
            final_datas.append({
                'id': len(final_datas),
                'input': question,
                'hint': hint,
                'requirement': requirement
            })
```

```python
id = 10
print(final_datas[id]['input'])
print("____"*1000)
print(final_datas[id]['hint'])
print("____"*1000)
print(final_datas[id]['requirement'])
print(len(final_datas))
```
