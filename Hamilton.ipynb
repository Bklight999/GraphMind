```python
import json
with open('./data_0104/Hamilton_path_true.json','r') as f:
    datas1 = json.load(f)
with open('./data_0104/Hamilton_path_false.json','r') as f:
    datas2 = json.load(f)

datas = datas1[:5000]+datas2[:10000]
print(len(datas))
```

```python
print(datas[0]['result'])
edges = datas[0]['edges']
edges = [tuple(edge) for edge in edges]
print(edges)
```

```python
final_datas = []
for data in datas:
    edges = data['edges']
    edges = [tuple(edge) for edge in edges]
    node_num = data['node_num']
    edges_str = ', '.join([f"({i}, {j})" for i, j in edges])
    question = f"""Determine whether or not there is a Hamiltonian path in an undirected graph. 
In an undirected graph, (i,j) means that node i and node j are connected with an undirected edge. 
Given a graph, you need to output Yes or No, indicating whether there is a Hamiltonian path in the graph.
Q: The nodes are numbered from 0 to {node_num-1}, and the edges are: {edges_str}. Is there a Hamiltonian path in this graph?"""

    answers = data['result']
    output = """"""
    if data['result'] == []:
        hint = f"Hint: There is no hamilton path in this graph.\n"
        requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [No].\n'
        final_datas.append({
            'input': question,
            'hint': hint,
            'requirement': requirement
        })
    else:
        for answer in answers:
            hint = f"Hint: A Hamilton path in this graph is as follows:"
            answer.append(answer[0])
            path = ' -> '.join([str(node) for node in answer])
            hint += f"\n{path}\n"
            requirement = 'Your output first contains your reasoning process and then pacakge your answer in the square, e.g. [Yes].\n'
            final_datas.append({
                'input': question,
                'hint': hint,
                'requirement': requirement
            })
```

```python
id = 10000
print(final_datas[id]['input'])
print("____"*1000)
print(final_datas[id]['hint'])
print("____"*1000)
print(final_datas[id]['requirement'])
print(len(final_datas))
```
